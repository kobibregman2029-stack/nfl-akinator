```python
# keep the game running until the player wants to stop 
playing = 'yes'
while playing == 'yes':
    # ask for the players name    
    name = input("What is your name: ")

    # easter egg for if aj/teacher plays the game
    if name.lower() == "aj":
        print("Hey AJ, if i can guess the NFL team your thinking of ima need that A+...")
    else:
        print("Hey " + name + ", Think of any NFL team for me to guess...")

    # asking to find out if the team plays in the NFC or AFC
    conference = input("Is your team in the NFC (yes/no):").lower()
    if conference == 'yes' or conference == 'nfc':
        # narrowing down the division
        division = input("Is your team in an Eastern division (yes/no):").lower()
        if division == 'yes' or division == 'east' or division == 'eastern':
            cowboys = input("Does your team have a star on their helmet (yes/no):").lower()
            if cowboys == 'yes':
                print("Your team is the Dallas Cowboys!")
            else:
                eagles = input("Does your team's logo have a bird on it (yes/no):").lower()
                if eagles == 'yes':
                    print("Your team is the Philadelphia Eagles!")
                else: 
                    giants = input("Does your team play at Metlife Stadium (yes/no):").lower()
                    if giants == 'yes':
                        print("Your team is the New York Giants!")
                    else:
                        commanders = input("Is your team the Washington Commanders (yes/no):").lower()
                        if commanders == 'yes':
                            print("Read your mind!")
                        else:
                            #default ending
                            print("I give up, what team were you thinking of?")
        else:
            print("This model only covers the NFC East ): Check out the completed version for all 32 NFL teams...")
    else:
        print("This model only covers the NFC East ): Check out the completed version for all 32 NFL teams...")
    
    playing = input("Want to play again (yes/no):").lower()

print("Thanks for playing!")
            
                
```
